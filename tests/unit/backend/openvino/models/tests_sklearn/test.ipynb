```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "ElasticNet", "Lasso", "LinearRegression", "NuSVR",
    "RandomForestRegressor", "Ridge", "SVR"
]
openvino = [0.2406, 0.1694, 0.0773, 25.6547, 40.6975, 0.1469, 23.1062]
sklearn = [1.860, 1.0225, 2.0135, 76.2563, 165.9936, 0.6541, 85.5864]
speedup = [7.7, 6.0, 25.7, 3.0, 4.1, 4.4, 3.7]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

rects1 = ax.bar(x - width/2, openvino, width, label='OpenVINO', color='#4CAF50')
rects2 = ax.bar(x + width/2, sklearn, width, label='Scikit-learn', color='#2196F3')
ax.set_ylabel('Execution Time (s) [log scale]')
ax.set_title('Performance Comparison: OpenVINO vs Scikit-learn')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=20)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.set_yscale('log')  # <-- Logarithmic scale

# Annotate speedup above bars
for i, s in enumerate(speedup):
    ax.text(x[i], max(openvino[i], sklearn[i]) * 1.2, f'{s:.1f}x', ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
